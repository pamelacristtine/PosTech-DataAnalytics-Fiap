
# **Tópico 1** - Top 10 dos países que mais tiveram crescimento na venda de vinhos de mesa | Valor exportado (US$))

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set_palette('Dark2')
palette = { 'purple', '#FFD700'}
```

```python
df_vinho = pd.read_csv('ExpVinho_2023.csv', delimiter=';', encoding='utf-8')
df_espumante = pd.read_csv('ExpEspumantes_2023.csv', delimiter=';', encoding='utf-8')
```

```python
df_vinho.columns
```

```python
df_vinho = df_vinho.drop('Id', axis=1)
```

```python
#df_vinho = df_vinho.set_index('País')
```

```python
#Valor Exportado -  Exportação Vinho(Últimos 15 anos)
colunas_qtd = ['País','2007.1','2008.1','2009.1','2010.1','2011.1','2012.1','2013.1','2014.1','2015.1','2016.1','2017.1','2018.1','2019.1','2020.1', '2021.1', '2022.1', '2023.1']
df_vinhos_filter_qtd = df_vinho.filter(colunas_qtd)
df_vinhos_filter_qtd.head()
```

```python
#Quantidade Exportado -  Exportação Vinho(Últimos 15 anos)
colunas_qtd = ['País','2008','2009','2010','2011','2012','2013','2014','2015','2016','2017','2018','2019','2020', '2021', '2022', '2023']
df_vinhos_filter_vol = df_vinho.filter(colunas_qtd)
df_vinhos_filter_vol.head()
```

```python
#Quantidade Exportado -  Exportação Espumante(Últimos 15 anos)
colunas_qtd = ['País','2008','2009','2010','2011','2012','2013','2014','2015','2016','2017','2018','2019','2020', '2021', '2022', '2023']
df_espumante_filter_vol = df_espumante.filter(colunas_qtd)
df_espumante_filter_vol.head()
```

```python
df_vinhos_filter_vol= df_vinhos_filter_vol.set_index('País').T
df_vinhos_filter_vol.head()
```

```python
df_vinhos_filter_qtd= df_vinhos_filter_qtd.set_index('País').T
df_vinhos_filter_qtd.head()
```

```python
df_espumante_filter_vol=  df_espumante_filter_vol.set_index('País').T
df_espumante_filter_vol.head()
```

```python
soma_por_pais = df_vinhos_filter_vol.sum(axis=0)
soma_por_pais= soma_por_pais
soma_por_pais = soma_por_pais.sort_values(ascending=False)
```

```python
soma_por_pais_espumante = df_espumante_filter_vol.sum(axis=0)
soma_por_pais_espumante= soma_por_pais_espumante
soma_por_pais_espumante = soma_por_pais_espumante.sort_values(ascending=False)
```

```python

dark_burgundy = '#800000'


# Gerar o gráfico de barras horizontais invertidas
plt.figure(figsize=(10, 6))
top_10_paises_soma = soma_por_pais.head(10).sort_values(ascending=True)
plt.barh(top_10_paises_soma.index, top_10_paises_soma.values, color= dark_burgundy)
plt.xlabel('Volume(L)')
plt.ylabel('País')
plt.grid(False)
plt.title('Maiores Importadores de Vinhos Mesa das Vinícolas do RS de 2008 a 2023')
plt.show()

```

```python

colunas_selecionadas = ['Paraguai', 'Rússia', 'Estados Unidos', 'China', 'Haiti']

df_colunas_selecionadas = df_vinhos_filter_qtd[colunas_selecionadas]
df_colunas_selecionadas
```

```python
df_vinhos_filter_qtd= df_colunas_selecionadas.T
df_vinhos_filter_qtd.head()
```

```python

df_crescimentos_percentuais = []

for pais, row in df_vinhos_filter_qtd.iterrows():
    for i in range(1, len(row)):
        ano_atual = df_vinhos_filter_qtd.columns[i]
        ano_anterior = df_vinhos_filter_qtd.columns[i - 1]

        valor_atual = row[ano_atual]
        valor_anterior = row[ano_anterior]

        if valor_anterior != 0:
            crescimento_percentual = ((valor_atual - valor_anterior) / valor_anterior) * 100
        else:
            crescimento_percentual = 0


        df_crescimentos_percentuais.append({"País": pais, "valor_atual":valor_atual, "Ano": ano_atual, "Variação": crescimento_percentual})


df_crescimentos_percentuais = pd.DataFrame(df_crescimentos_percentuais)

display(df_crescimentos_percentuais.head())
```

```python
crescimento_anos = df_crescimentos_percentuais[df_crescimentos_percentuais['Ano'].isin(['2013.1','2014.1','2015.1','2016.1','2017.1','2018.1','2019.1','2020.1','2021.1','2022.1', '2023.1'])].reset_index()
crescimento_anos.head()
```

```python

df_media_crescimento_pais = crescimento_anos .groupby('País')['Variação'].mean().reset_index()

df_top_10_paises = df_media_crescimento_pais.sort_values(by='Variação', ascending=False)

top_10_paises = df_top_10_paises.head(10)

print("Top 10 Países com Maior Crescimento Percentual Médio:")
print(top_10_paises)
```

```python
# Os 10 países com maior maior crescimento percenteual médio
purple_light_rgb = (211/255, 209/255, 227/255)
purple_dark_rgb = (93/255, 50/255, 115/255)

light_burgundy = '#A52A2A'
dark_burgundy = '#800000'

colors = [dark_burgundy if i < 2 else light_burgundy for i in range(len(top_10_paises))]


plt.figure(figsize=(10, 6))
plt.bar(top_10_paises['País'], top_10_paises['Variação'], color=colors)
plt.xlabel('País')
plt.ylabel('Crescimento Percentual (%)')
plt.title("Crescimento Percentual Médio nas Exportações de Vinho de Mesa das Vinícolas do RS (US$)")
#plt.xticks(rotation=45)
plt.grid(False)
plt.show()
```

```python
crescimento_venda = df_crescimentos_percentuais[df_crescimentos_percentuais['País'].isin(['China','Rússia', 'Haiti','Paraguai', 'Estados Unidos'])].reset_index()
```

```python
crescimento_venda = crescimento_venda.drop(columns=['index'])
crescimento_venda. columns
```

```python
crescimento_venda['Ano'] =crescimento_venda['Ano'].str.replace('.1', '')
```

```python
crescimento_venda.head()
```

```python
lightest_burgundy = '#FFB6C1'
salmon_color = '#FA8072'
marrom_claro= '#D2B48C'
marrom_escuro = '#8B4513'

plt.figure(figsize=(10, 6))

colors = [dark_burgundy, salmon_color, marrom_claro, lightest_burgundy, marrom_escuro]

for i, pais in enumerate(crescimento_venda['País'].unique()):
    dt = crescimento_venda[crescimento_venda['País'] == pais]
    plt.plot(dt['Ano'], dt['valor_atual'], marker='o', label=pais, color=colors[i])

plt.xlabel('Ano')
plt.ylabel('Valor Exportado (US$)')
plt.title('Exportações de Vinho de Mesa das Vínicolas do RS (2008-2023)')

plt.legend()
plt.grid(False)
plt.show()
```

#**Tópico 2** - Tendência de vendas por tipo de produto (vinho de mesa e espumante) | Comercialização

```python
df_comercio = pd.read_csv('/content/Comercio.csv', delimiter=';', encoding='utf-8')
df_comercio.head(5)
```

```python
### BASE COM REPARTIÇÃO POR GRUPO DE CADA PRODUTO

# Máscara booleana para os IDs desejados
ids_desejados = list(range(1, 9)) + list(range(15, 19))
mask = df_comercio['id'].isin(ids_desejados)

df_filtrado_com = df_comercio[mask]

display(df_filtrado_com.head())
```

```python
### BASE COM REPARTIÇÃO GERAL

ids_desejados = [1, 15]
mask = df_comercio['id'].isin(ids_desejados)

df_filtrado_geral = df_comercio[mask]

display(df_filtrado_geral.head())
```

```python
df_comercio_geral = df_filtrado_geral.drop(['id', 'control'], axis=1)
```

```python
df_comercio_geral.head()
```

```python
#Tendencia de Vendas- Comercialização (Últimos 15 anos)
colunas_filter = ['Produto','2008','2009','2010','2011','2012','2013','2014','2015','2016','2017','2018','2019','2020', '2021', '2022', '2023']
df_comercio_geral_filter = df_comercio_geral.filter(colunas_filter)
df_comercio_geral_filter = df_comercio_geral_filter.reset_index(drop=True)
df_comercio_geral_filter
```

```python

dark_burgundy = "#6E0D1A"
marrom_claro = "#C69C6D"


df = df_comercio_geral_filter.T
df.columns = df.iloc[0]
df = df[1:]
df = df.apply(pd.to_numeric)
df.reset_index(inplace=True)
df.rename(columns={'index': 'Ano'}, inplace=True)
df['Ano'] = df['Ano'].astype(int)

df_long = df.melt(id_vars='Ano', var_name='Produto', value_name='Quantidade Vendida')

sns.set_style("whitegrid")
plt.figure(figsize=(10, 6))

unique_products = df_long['Produto'].unique()
palette = {product: color for product, color in zip(unique_products, [dark_burgundy, marrom_claro])}

sns.lineplot(data=df_long, x='Ano', y='Quantidade Vendida', hue='Produto', marker='o', palette=palette)
plt.xlabel('Ano')
plt.ylabel('Quantidade Vendida (L)')
plt.title('Tendência de Vendas por Tipo de Produto (2008-2023)')

plt.legend(title='Tipo de Produto', loc='best')

min_year = df_long['Ano'].min() - 1
max_year = df_long['Ano'].max() + 1
plt.xlim(min_year, max_year)

plt.xticks(df_long['Ano'].unique(), rotation=45)

plt.grid(False)
plt.show()
```

#**Tópico 3** - Top 10 do Maior Preço médio por litro - Espumante (litros / receita)
# Razão Valor-por-Quantidade

```python
df_espumante.head(5)
```

```python
colunas_esp = ['País','2008','2009','2010','2011','2012','2013','2014','2015','2016','2017','2018','2020', '2021', '2022', '2023',
               '2008.1','2009.1','2010.1','2011.1','2012.1','2013.1','2014.1','2015.1','2016.1','2017.1','2018.1','2020.1', '2021.1', '2022.1', '2023.1']
df_espumante_filter = df_espumante.filter(colunas_esp)
df_espumante_filter.head()
```

```python
df_espumante_filter.columns
```

```python

preco_medio_por_pais = []

for index, row in df_espumante_filter.iterrows():
    pais = row['País']
    receita = pd.to_numeric(row.iloc[1::2], errors='coerce')
    volume_vendido = pd.to_numeric(row.iloc[::2], errors='coerce')
    preco_medio = receita.sum() / volume_vendido.sum()
    preco_medio_por_pais.append((pais, preco_medio))

df_preco_medio_por_pais = pd.DataFrame(preco_medio_por_pais, columns=['País', 'Preço Médio por Litro'])

print(df_preco_medio_por_pais.head())
```

```python
#Top 10 países com o maior preço médio por litro
top_10_paises = df_preco_medio_por_pais.sort_values(by='Preço Médio por Litro', ascending=False).head(10)
print(top_10_paises)
```

```python

#Top 10 Países com Maior Preço Médio por Litro de Espumante - Análise realizada considerando os últimos 15 anos (2008-2023)

top_10_paises = df_preco_medio_por_pais.sort_values(by='Preço Médio por Litro', ascending=False).head(10)

num_barras = len(top_10_paises)

palette = sns.light_palette(dark_burgundy, n_colors=num_barras, reverse=True)

plt.figure(figsize=(10, 6))
grafico = sns.barplot(x='Preço Médio por Litro', y='País', data=top_10_paises, palette=palette, orient='h')

for i in range(num_barras):
    valor = top_10_paises.iloc[i]['Preço Médio por Litro']
    grafico.text(valor, i, f'{valor:.2f}', va='center', fontweight='bold')

espaco_extra = -0.2
ymin, ymax = plt.ylim()
plt.ylim(ymin - espaco_extra, ymax + espaco_extra)


plt.xlabel('Preço Médio por Litro (US$)')
plt.ylabel('País')
plt.title('Top 10 Países com Maior Preço Médio por Litro de Espumante (2008-2023)')
plt.tight_layout()
plt.grid(False)
plt.show()

```

```python
colunas_esp = ['País','2008','2009','2010','2011','2012','2013','2014','2015','2016','2017','2018','2020', '2021', '2022', '2023',
               '2008.1','2009.1','2010.1','2011.1','2012.1','2013.1','2014.1','2015.1','2016.1','2017.1','2018.1','2020.1', '2021.1', '2022.1', '2023.1']
df_vinho_filter = df_vinho.filter(colunas_esp)
df_vinho_filter.head()
```

```python

preco_medio_por_pais = []

for index, row in df_vinho_filter.iterrows():
    pais = row['País']
    receita = pd.to_numeric(row.iloc[1::2], errors='coerce')
    volume_vendido = pd.to_numeric(row.iloc[::2], errors='coerce')

    if volume_vendido.sum() == 0:
        preco_medio = 0
    else:
        preco_medio = receita.sum() / volume_vendido.sum()

    preco_medio_por_pais.append((pais, preco_medio))

df_preco_medio_por_pais_vinho = pd.DataFrame(preco_medio_por_pais, columns=['País', 'Preço Médio por Litro'])

# Preço médio por litro para cada país
```

```python
# Top 10 países com o maior preço médio por litro
top_10_paises_vinho = df_preco_medio_por_pais_vinho.sort_values(by='Preço Médio por Litro', ascending=False).head(10)
print(top_10_paises_vinho)
```

```python
#Top 10 Países com Maior Preço Médio por Litro de Vinho de Mesa - Análise realizada considerando os últimos 15 anos (2008-2023)
num_barras = len(top_10_paises_vinho)

palette = sns.light_palette(dark_burgundy, n_colors=num_barras, reverse=True)

plt.figure(figsize=(10, 6))
grafico = sns.barplot(x='Preço Médio por Litro', y='País', data=top_10_paises_vinho, palette=palette, orient='h')

for i in range(num_barras):
    valor = top_10_paises_vinho.iloc[i]['Preço Médio por Litro']
    grafico.text(valor, i, f'{valor:.2f}', va='center', fontweight='bold')

espaco_extra = -0.2
ymin, ymax = plt.ylim()
plt.ylim(ymin - espaco_extra, ymax + espaco_extra)


plt.xlabel('Preço Médio por Litro (US$)')
plt.ylabel('País')
plt.title('Top 10 Países com Maior Preço Médio por Litro de Vinho de Mesa (2008-2023) ')
plt.tight_layout()
plt.grid(False)
plt.show()
```
